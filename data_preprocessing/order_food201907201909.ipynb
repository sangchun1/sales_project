```python
# 필요한 모듈 불러오기
import pandas as pd
import numpy as np
```

```python
# 데이터 파일 불러오기
df = pd.read_csv("./KGU_3rd_ORIGIN_KGUAREADOITEM_20190701000000 (1).csv")
```

```python
df
```

```python
# 데이터 타입 확인
df.info()
```

```python
# 날짜 타입을 datetime으로 변환
df['2019-07-31'] = pd.to_datetime(df['2019-07-31'])
```

```python
df.info()
```

```python
# 결측치 확인
df.isnull().sum()
```

```python
# 컬럼명 변경
df.rename(columns={'2019-07-31':'date'}, inplace=True)
df.rename(columns={'00':'time_int'}, inplace=True)
df.rename(columns={'경기도':'city'},inplace=True)
df.rename(columns={'수요일':'days'},inplace=True)
df.rename(columns={'고양시 일산서구':'gu'},inplace=True)
df.rename(columns={'피자':'food'},inplace=True)
```

```python
df
```

```python
# 시도에서 서울특별시에 해당하는 데이터만 추출
df_seoul = df.loc[df['city'].isin(['서울특별시'])]
```

```python
# 인덱스 초기화
df_seoul.reset_index(drop=True, inplace=True)
```

```python
df_seoul
```

```python
# 필요없는 데이터 열 삭제
df_seoul.drop(['city'],axis=1,inplace=True)
```

```python
df_seoul
```

```python
# 데이터 타입 확인
df_seoul.info()
```

```python
# csv로 저장
df_seoul.to_csv("201907~201909지역별요일별배달품목.csv")
```
