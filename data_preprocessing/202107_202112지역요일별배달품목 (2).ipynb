```python
# 필요한 모듈 불러오기
import pandas as pd
import numpy as np
```

```python
# 데이터 파일 불러오기
df = pd.read_csv("./KGU_3rd_ORIGIN_KGUAREADOITEM_20210701000000.csv")
```

```python
df
```

```python
df.info()
```

```python
df['﻿2021-07-01'] = pd.to_datetime(df['﻿2021-07-01'])
```

```python
df.info()
```

```python
# 결측치 확인
df.isnull().sum()
```

```python
# 컬럼명 변경
df.rename(columns={'﻿2021-07-01':'년월일'}, inplace=True)
df.rename(columns={'00':'주문시간'}, inplace=True)
df.rename(columns={'충청남도':'시도'},inplace=True)
df.rename(columns={'목요일':'요일'},inplace=True)
df.rename(columns={'천안시 동남구':'시군구'},inplace=True)
df.rename(columns={'돈까스/일식':'음식분류'},inplace=True)
```

```python
df
```

```python
# 시도에서 서울특별시에 해당하는 데이터만 추출
df_seoul = df.loc[df['시도'].isin(['서울특별시'])]
```

```python
df_seoul.reset_index()
df_seoul.reset_index(drop=True, inplace=True)
```

```python
df_seoul
```

```python
df_seoul = df_seoul[df_seoul.음식분류 != '심부름']
df_seoul = df_seoul[df_seoul.음식분류 != '야식']
df_seoul = df_seoul[df_seoul.음식분류 != '배달전문업체']
```

```python
df_seoul
```

```python
df_seoul.to_csv("202107~202112지역별요일별배달품목.csv")
```
