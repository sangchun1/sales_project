```python
# 필요한 모듈 불러오기
import pandas as pd
import numpy as np
```

```python
# 데이터 파일 불러오기
df = pd.read_csv("./KGU_3rd_ORIGIN_KGUAREADOITEM_20200101000000.csv")
```

```python
df.info()
```

```python
df['2020-01-01'] = pd.to_datetime(df['2020-01-01'])
```

```python
# 결측치 확인
df.isnull().sum()
```

```python
# 컬럼명 변경
df.rename(columns={'2020-01-01':'date'}, inplace=True)
df.rename(columns={'00':'time_int'}, inplace=True)
df.rename(columns={'경상남도':'city'},inplace=True)
df.rename(columns={'수요일':'days'},inplace=True)
df.rename(columns={'창원시 마산회원구':'gu'},inplace=True)
df.rename(columns={'치킨':'food'},inplace=True)
```

```python
# 시도에서 서울특별시에 해당하는 데이터만 추출
df_seoul = df.loc[df['city'].isin(['서울특별시'])]
```
