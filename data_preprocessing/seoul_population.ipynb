# 1. 데이터 불러오기

```python
# 필요한 모듈 불러오기
import pandas as pd
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning) # 경고 메세지 무시
```

```python
# csv 파일 불러오기
df = pd.read_csv('D:/박상춘/프로젝트/raw data/서울시 행정 데이터/연령_및_성별_인구.csv', header=1, encoding='cp949')
```

# 2. 데이터 확인 및 전처리

```python
# 데이터 확인
df.head(25)
```

```python
# 컬럼명 변경
df.rename({'행정구역별(읍면동)':'gu', '연령별':'age_group', '총인구(명)':'population', '총인구_남자(명)':'male', '총인구_여자(명)':'female'}, axis=1, inplace=True)
```

```python
# 서울시 합계 데이터 제거
df = df[df.gu != '서울특별시']
```

```python
df.head(20)
```

```python
# 인덱스 리셋
df.set_index([pd.Index(range(0,df.gu.count()))], inplace=True)
```

```python
gu_arr = df.gu.unique()
gu_lis = gu_arr.tolist()
```

```python
# 연령대 카테고리 축소
lis1, lis2, lis3, lis4, lis5, lis6 = [],[],[],[],[],[]
df1 = pd.DataFrame(columns=['gu', 'age_group', 'population', 'male', 'female'])
for gu in gu_lis: # 자치구별로
    df_t = df[df['gu'] == gu] # 임시 데이터 프레임에 저장
        
    lis1 = [gu, '20대 미만', df_t.iloc[0:4, 2].sum(), df_t.iloc[0:4, 3].sum(), df_t.iloc[0:4, 4].sum()] # 20대 미만
    lis2 = [gu, '20대', df_t.iloc[4:6, 2].sum(), df_t.iloc[4:6, 3].sum(), df_t.iloc[4:6, 4].sum()]      # 20대
    lis3 = [gu, '30대', df_t.iloc[6:7, 2].sum(), df_t.iloc[6:7, 3].sum(), df_t.iloc[6:7, 4].sum()]      # 30대
    lis4 = [gu, '40대', df_t.iloc[7:8, 2].sum(), df_t.iloc[7:8, 3].sum(), df_t.iloc[7:8, 4].sum()]      # 40대
    lis5 = [gu, '50대', df_t.iloc[8:9, 2].sum(), df_t.iloc[8:9, 3].sum(), df_t.iloc[8:9, 4].sum()]      # 50대
    lis6 = [gu, '60대 이상', df_t.iloc[9:, 2].sum(), df_t.iloc[9:, 3].sum(), df_t.iloc[9:, 4].sum()]    # 60대 이상

    df1 = df1.append(pd.Series(lis1, index=df1.columns), ignore_index=True) # 20대 미만
    df1 = df1.append(pd.Series(lis2, index=df1.columns), ignore_index=True) # 20대
    df1 = df1.append(pd.Series(lis3, index=df1.columns), ignore_index=True) # 30대
    df1 = df1.append(pd.Series(lis4, index=df1.columns), ignore_index=True) # 40대
    df1 = df1.append(pd.Series(lis5, index=df1.columns), ignore_index=True) # 50대
    df1 = df1.append(pd.Series(lis6, index=df1.columns), ignore_index=True) # 60대 이상
```
