```python
import pandas as pd
```

```python
df = pd.read_csv("KGU_2nd_ORIGIN_KGUWETHERINFO_20201031000000.csv")
```

```python
df
```

```python
df.info()
```

```python
# 날짜 타입을 datetime으로 변환
df['20190101'] = pd.to_datetime(df['20190101'])
```

```python
df.info()
```

```python
# 필요없는 데이터 삭제
df.drop(['1'],axis=1,inplace=True)
df.drop(['SW'],axis=1,inplace=True)
df.drop(['228'],axis=1,inplace=True)
df.drop(['east'],axis=1,inplace=True)
df.drop(['42150'],axis=1,inplace=True)
df.drop(['0000'],axis=1,inplace=True)
df.drop(['없음'],axis=1,inplace=True)
df.drop(['25'],axis=1,inplace=True)
df.drop(['약간강'],axis=1,inplace=True)
```

```python
df
```

```python
# 컬럼명 변경
df.rename(columns={'20190101':'date'}, inplace=True)
df.rename(columns={'강원도':'city'}, inplace=True)
df.rename(columns={'강릉시':'gu'},inplace=True)
df.rename(columns={'0':'pph'},inplace=True)
df.rename(columns={'-0.3':'temp'},inplace=True)
df.rename(columns={'4.5':'ws'},inplace=True)
```

```python
df
```

```python
# 시도에서 서울특별시에 해당하는 데이터만 추출
df_seoul = df.loc[df['city'].isin(['서울특별시'])]
```

```python
df_seoul
```

```python
# 인덱스 초기화
df_seoul.reset_index(drop=True, inplace=True)
```

```python
df_seoul
```

```python
# 필요없는 데이터 열 삭제
df_seoul.drop(['city'],axis=1,inplace=True)
```

```python
df_seoul
```

```python
# 데이터 타입 확인
df_seoul.info()
```

```python
df_seoul.isnull().sum()
```

```python
# csv로 저장
df_seoul.to_csv("201901~201912 Order area weather information.csv")
```
