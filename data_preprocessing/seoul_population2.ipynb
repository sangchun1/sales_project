```python
# 연령대 축소
for i in df.index:
    # 남자
    df.loc[i, '남자20대미만'] = df.iloc[i, 4:7].sum()
    df.loc[i, '남자20대'] = df.iloc[i, 7:9].sum()
    df.loc[i, '남자30대'] = df.iloc[i, 9:11].sum()
    df.loc[i, '남자40대'] = df.iloc[i, 11:13].sum()
    df.loc[i, '남자50대'] = df.iloc[i, 13:15].sum()
    df.loc[i, '남자60대이상'] = df.iloc[i, 15:18].sum()
    # 여자
    df.loc[i, '여자20대미만'] = df.iloc[i, 18:21].sum()
    df.loc[i, '여자20대'] = df.iloc[i, 21:23].sum()
    df.loc[i, '여자30대'] = df.iloc[i, 23:25].sum()
    df.loc[i, '여자40대'] = df.iloc[i, 25:27].sum()
    df.loc[i, '여자50대'] = df.iloc[i, 27:29].sum()
    df.loc[i, '여자60대이상'] = df.iloc[i, 29:32].sum()
```

```python
# 연령대별 남녀 합계 계산
for i in df.index:
    df.loc[i, '20대미만'] = df.loc[i, '남자20대미만'] + df.loc[i, '여자20대미만']
    df.loc[i, '20대'] = df.loc[i, '남자20대'] + df.loc[i, '여자20대']
    df.loc[i, '30대'] = df.loc[i, '남자30대'] + df.loc[i, '여자30대']
    df.loc[i, '40대'] = df.loc[i, '남자40대'] + df.loc[i, '여자40대']
    df.loc[i, '50대'] = df.loc[i, '남자50대'] + df.loc[i, '여자50대']
    df.loc[i, '60대이상'] = df.loc[i, '남자60대이상'] + df.loc[i, '여자60대이상']
```

```python
# 성별 합계 계산
for i in df.index:
    df.loc[i, 'male_total'] = df.loc[i, '남자20대미만':'남자60대이상'].sum()
    df.loc[i, 'female_total'] = df.loc[i, '여자20대미만':'여자60대이상'].sum()
```
