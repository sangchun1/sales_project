```python
# 필요한 모듈 불러오기
import pandas as pd
import numpy as np
```

```python
# 데이터 파일 불러오기
df = pd.read_csv("./KGU_3rd_ORIGIN_KGUAREADOITEM_20210101000000.csv")
```

```python
df
```

```python
# 결측치 확인
df.isnull().sum()
```

```python
# 컬럼명 변경
df.rename(columns={'﻿2021-01-01':'년월일'}, inplace=True)
df.rename(columns={'00':'주문시간'}, inplace=True)
df.rename(columns={'경기도':'시도'},inplace=True)
df.rename(columns={'금요일':'요일'},inplace=True)
df.rename(columns={'의정부시':'시군구'},inplace=True)
df.rename(columns={'배달전문업체':'음식분류'},inplace=True)
```

```python
df
```

```python
# 시도에서 서울특별시에 해당하는 데이터만 추출
df_seoul = df.loc[df['시도'].isin(['서울특별시'])]
```

```python
# 인덱스 기준을 년월일로 수정
df_seoul.set_index('년월일', inplace=True)
```

```python
df_seoul
```

```python
df_seoul = df_seoul[df_seoul.음식분류 != '심부름']
df_seoul = df_seoul[df_seoul.음식분류 != '야식']
df_seoul = df_seoul[df_seoul.음식분류 != '배달전문업체']
```

```python
df_seoul
```

```python
df_seoul.to_csv("202101~202106지역별요일별배달품목.csv")
```
