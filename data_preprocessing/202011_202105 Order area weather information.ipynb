```python
import pandas as pd
```

```python
df = pd.read_csv("KGU_2nd_ORIGIN_KGUWETHERINFO_20211013000000.csv")
```

```python
df
```

```python
df.info()
```

```python
df['20201109']= df['20201109'].astype('str')
```

```python
# 날짜 타입을 datetime으로 변환
df['20201109'] = pd.to_datetime(df['20201109'])
```

```python
df.info()
```

```python
# 필요없는 데이터 삭제
df.drop(['4020001'],axis=1,inplace=True)
df.drop(['11110'],axis=1,inplace=True)
df.drop(['0000'],axis=1,inplace=True)
df.drop(['없음'],axis=1,inplace=True)
df.drop(['49'],axis=1,inplace=True)
df.drop(['약'],axis=1,inplace=True)
df.drop(['east'],axis=1,inplace=True)
df.drop(['350'],axis=1,inplace=True)
df.drop(['N'],axis=1,inplace=True)
```

```python
df
```

```python
# 컬럼명 변경
df.rename(columns={'20201109':'date'}, inplace=True)
df.rename(columns={'서울특별시':'city'}, inplace=True)
df.rename(columns={'종로구':'gu'},inplace=True)
df.rename(columns={'0':'pph'},inplace=True)
df.rename(columns={'2.5':'temp'},inplace=True)
df.rename(columns={'0.8':'ws'},inplace=True)
```

```python
df
```

```python
# 시도에서 서울특별시에 해당하는 데이터만 추출
df_seoul = df.loc[df['city'].isin(['서울특별시'])]
```

```python
df_seoul
```

```python
# 인덱스 초기화
df_seoul.reset_index(drop=True, inplace=True)
```

```python
df_seoul
```

```python
# 필요없는 데이터 열 삭제
df_seoul.drop(['city'],axis=1,inplace=True)
```

```python
df_seoul
```

```python
# 데이터 타입 확인
df_seoul.info()
```
