# 1. 데이터 불러오기

```python
# 필요한 모듈 불러오기
import pandas as pd
```

```python
# csv 파일을 하나씩 데이터 프레임으로 변환
df1 = pd.read_csv('D:/박상춘/프로젝트/raw data/시간대별 주문금액 데이터(연월일, 주문시간, 업종, 지역(시도), 주문금액)/KGU_3rd_ORIGIN_KGUDSTNAMOUNT_2019_05~12.csv', header=None)
df2 = pd.read_csv('D:/박상춘/프로젝트/raw data/시간대별 주문금액 데이터(연월일, 주문시간, 업종, 지역(시도), 주문금액)/KGU_3rd_ORIGIN_KGUDSTNAMOUNT_20200101000000.csv', header=None)
df3 = pd.read_csv('D:/박상춘/프로젝트/raw data/시간대별 주문금액 데이터(연월일, 주문시간, 업종, 지역(시도), 주문금액)/KGU_3rd_ORIGIN_KGUDSTNAMOUNT_2021_01~07.csv', header=None)
df4 = pd.read_csv('D:/박상춘/프로젝트/raw data/시간대별 주문건수 데이터(연월일, 주문시간, 업종, 지역(시도), 주문건수)/KGU_3rd_ORIGIN_KGUDSTNORDER_20190101000000.csv', header=None)
df5 = pd.read_csv('D:/박상춘/프로젝트/raw data/시간대별 주문건수 데이터(연월일, 주문시간, 업종, 지역(시도), 주문건수)/KGU_3rd_ORIGIN_KGUDSTNORDER_20200101000000.csv', header=None)
df6 = pd.read_csv('D:/박상춘/프로젝트/raw data/시간대별 주문건수 데이터(연월일, 주문시간, 업종, 지역(시도), 주문건수)/KGU_3rd_ORIGIN_KGUDSTNORDER_20210101000000.csv', header=None)
```

```python
# 데이터 프레임을 하나로 합치기
df_amt = pd.concat([df1, df2, df3])
df_cnt = pd.concat([df4, df5, df6])
```

# 2. 데이터 확인 및 전처리

```python
df_amt
```

```python
df_cnt
```

```python
# 두 데이터 프레임 합치기
df = pd.merge(left=df_amt, right=df_cnt, how='inner', on=[0, 1, 2, 3, 4])
```

```python
# 컬럼명 바꾸기
df.rename({0:'date', 1:'time_int', 2:'food', 3:'city', 4:'gu', '5_x':'avg_order_amt', '5_y':'order_cnt'}, axis=1, inplace=True)
```

```python
# 데이터 타입 확인
print(df.dtypes)
```

```python
# date 컬럼을 시간 데이터 타입으로 변경
df['date'] = pd.to_datetime(df['date'])
```

```python
# 서울시 데이터만 추출
df = df[df.city == '서울특별시']
```

```python
# 불필요한 도시 컬럼 제거
df.drop(['city'], axis=1, inplace=True)
```

```python
# 결측치 확인
df.isnull().sum()
```

```python
# 인덱스 리셋
df.set_index([pd.Index(range(0,741432))], inplace=True)
```
