### 4. 데이터중 불필요한 데이터 삭제

```python
result_seoul_drop = result_seoul.loc[(result_seoul['food'] != '배달전문업체') & (result_seoul['food'] != '테스트') & (result_seoul['food'] != '심부름')]
```

### 6. 가게별 위도, 경도 추가

```python
# 카카오 API KEY 불러오기
with open('C:/Kcoding/KakaoAPI.txt', "r") as f:
    api_key = f.read()
```

```python
# 잘 불러와졌는지 확인
len(api_key)
```

```python
result_del['address']
```

```python
# 첫번째 데이터 샘플링
addr = '서울특별시 강북구 솔샘로 (미아동)'
quote(addr)
search_url = "https://dapi.kakao.com/v2/local/search/address.json"
url = f'{search_url}?query={quote(addr)}'
result = requests.get(url, headers = {'Authorization': f'KakaoAK {api_key}'}).json()
result
```

```python
lat_list, lng_list = [], []

for i in tqdm(result_del['address']):
    addr = f'{(i)}'
    quote(addr)
    search_url = "https://dapi.kakao.com/v2/local/search/address.json"
    url = f'{search_url}?query={quote(addr)}'
    result = requests.get(url, headers = {'Authorization': f'KakaoAK {api_key}'}).json()
    lng = float(result['documents'][0]['x'])
    lat = float(result['documents'][0]['y'])
    lat_list.append(lat)
    lng_list.append(lng)
```

```python
# 위도, 경도 데이터 추가하기
result_del['lat'] = lat_list
result_del['lng'] = lng_list
```
