```python
# 필요한 모듈 불러오기
import pandas as pd
import numpy as np
```

```python
# 데이터 파일 불러오기
df = pd.read_csv("./KGU_3rd_ORIGIN_KGUAREADOITEM_20190701000000 (1).csv")
```

```python
df
```

```python
df.info()
```

```python
df['2019-07-31'] = pd.to_datetime(df['2019-07-31'])
```

```python
df.info()
```

```python
# 결측치 확인
df.isnull().sum()
```

```python
# 컬럼명 변경
df.rename(columns={'2019-07-31':'년월일'}, inplace=True)
df.rename(columns={'00':'주문시간'}, inplace=True)
df.rename(columns={'경기도':'시도'},inplace=True)
df.rename(columns={'수요일':'요일'},inplace=True)
df.rename(columns={'고양시 일산서구':'시군구'},inplace=True)
df.rename(columns={'피자':'음식분류'},inplace=True)
```

```python
df
```

```python
# 시도에서 서울특별시에 해당하는 데이터만 추출
df_seoul = df.loc[df['시도'].isin(['서울특별시'])]
```

```python
df_seoul.reset_index()
```

```python
df_seoul.reset_index(drop=True, inplace=True)
```

```python
df_seoul = df_seoul[df_seoul.음식분류 != '심부름']
```

```python
df_seoul = df_seoul[df_seoul.음식분류 != '야식']
```

```python
df_seoul = df_seoul[df_seoul.음식분류 != '배달전문업체']
```
